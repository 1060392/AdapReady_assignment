```python
#1
import pandas as pd

data_url = "https://files.consumerfinance.gov/ccdb/complaints.csv.zip"
complaints_df = pd.read_csv(data_url, compression='zip')

print("First few rows of the DataFrame:")
print(complaints_df.head())

print("\nDataFrame information:")
print(complaints_df.info())

print("\nBasic statistics of numerical columns:")
print(complaints_df.describe())

product_counts = complaints_df['Product'].value_counts()
print("\nCount of complaints by product:")
print(product_counts)

company_counts = complaints_df['Company'].value_counts().head(10)
print("\nCount of complaints by company (Top 10):")
print(company_counts)
```

```python
#2
def findLengthOfLCIS(nums):
    if not nums:
        return 0
    
    max_length = 1
    current_length = 1
    
    for i in range(1, len(nums)):
        if nums[i] > nums[i - 1]:
            current_length += 1
            max_length = max(max_length, current_length)
        else:
            current_length = 1
    
    return max_length

# Example usage:
nums1 = [1, 3, 5, 4, 7]
nums2 = [2, 2, 2, 2, 2]

print("Example 1 Output:", findLengthOfLCIS(nums1))
print("Example 2 Output:", findLengthOfLCIS(nums2))  
```

```python
#4
import json
import csv

with open('sample_json.json', 'r') as json_file:
    data = json.load(json_file)

servlets = data['web-app']['servlet']

csv_file = 'servlet_info.csv'

with open(csv_file, mode='w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(['servlet-name', 'servlet-class'])
    for servlet in servlets:
        servlet_name = servlet['servlet-name']
        servlet_class = servlet['servlet-class']
        writer.writerow([servlet_name, servlet_class])

print(f"Data has been stored in '{csv_file}'")
```
